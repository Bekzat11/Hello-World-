```python
def stemmer(text):
  # -- first clear string from symbols and split each word
  text = text.split(' ')
  punctuation='~!@#$%^&*()-_+=:;\'\",./?«»'  
  de_punct_sent=[]
  for s in text:
      for p in punctuation:
          s=s.replace(p,'')
      s=s.replace('  ', '')
      s= s.lower()
      de_punct_sent.append(s)

  split_text = de_punct_sent
  count =0
  max =0
  mmax=0
  new_text ={}
  # -- compare each substrings with all words in text and find optimal substring to define as root
  for i in range(len(split_text)):
    mmax =0
    for l in range(2,len(split_text[i])):
      root = split_text[i][:l+1]
      count=0
      if mmax>2:
        max = int(mmax/2)
      else:
        max =0
      for t in range(len(split_text)):
        if t!=i:
          if root in split_text[t]:
            count +=1
      if count>max:
        new_text[split_text[i]]=root
        max =count
        if max >mmax:
          mmax =max
    if mmax ==0:
      new_text[split_text[i]]=split_text[i]
    if len(new_text[split_text[i]])<2:
      new_text[split_text[i]]=split_text[i]
  # -- return a dictionary with words and their roots
  return new_text
```
